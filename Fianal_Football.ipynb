```python
import pandas as pd
import plotly.graph_objects as go

# Supongamos que tienes una matriz de correlación llamada 'correlation' y un DataFrame llamado 'correlation_df' que contiene los valores

# Configura las etiquetas de los ejes X e Y
x_labels = correlation_pl.columns
y_labels = correlation_pl.index

# Crea una matriz de anotaciones para mostrar los valores de correlación en cada celda
annotations = []
for y, ylabel in enumerate(y_labels):
    for x, xlabel in enumerate(x_labels):
        annotations.append(
            dict(
                x=x,
                y=y,
                text=str(round(correlation_pl.loc[ylabel, xlabel], 2)),
                showarrow=False,
                font=dict(color='deepskyblue'),
            )
        )

# Crea la figura del heatmap
fig = go.Figure(
    data=go.Heatmap(
        z=correlation_pl,
        x=x_labels,
        y=y_labels,
        colorscale='Blues',
        colorbar=dict(title='Correlation'),
    )
)

# Agrega las anotaciones a la figura
fig.update_layout(
    annotations=annotations,
    xaxis=dict(tickangle=20),
    yaxis=dict(tickangle=0),
    title='Correlation Heatmap',
)

fig.show()
```

```python
# This is the General analysis by Premier league Team

# Selecting the relevant columns ( all of them ) 
columnas = ['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', 'MP', 'Starts', 'Min', '90s', 'Goals', 'Shots', 'SoT%', 'G/Sh', 
            'G/SoT', 'ShoDist', 'ShoFK', 'ShoPK', 'PKatt', 'PasTotCmp', 'PasTotAtt', 'PasTotCmp%', 'PasTotDist', 'PasTotPrgDist',
            'PasShoCmp', 'PasShoAtt', 'PasShoCmp%', 'PasMedCmp', 'PasMedAtt', 'PasMedCmp%', 'PasLonCmp', 'PasLonAtt', 'PasLonCmp%',
            'Assists', 'PasAss', 'Pas3rd', 'PPA', 'CrsPA', 'PasProg', 'PasAtt', 'PasLive', 'PasDead', 'PasFK', 'TB', 'Sw', 
            'PasCrs', 'TI', 'CK', 'CkIn', 'CkOut', 'CkStr', 'PasCmp', 'PasOff', 'PasBlocks', 'SCA', 'ScaPassLive', 'ScaPassDead',
            'ScaDrib', 'ScaSh', 'ScaFld', 'ScaDef', 'GCA', 'GcaPassLive', 'GcaPassDead', 'GcaDrib', 'GcaSh', 'GcaFld', 'GcaDef', 
            'Tkl', 'TklWon', 'TklDef3rd', 'TklMid3rd', 'TklAtt3rd', 'TklDri', 'TklDriAtt', 'TklDri%', 'TklDriPast', 'Blocks', 'BlkSh', 
            'BlkPass', 'Int', 'Tkl+Int', 'Clr', 'Err', 'Touches', 'TouDefPen', 'TouDef3rd', 'TouMid3rd', 'TouAtt3rd', 'TouAttPen', 
            'TouLive', 'ToAtt', 'ToSuc', 'ToSuc%', 'ToTkl', 'ToTkl%', 'Carries', 'CarTotDist', 'CarPrgDist', 'CarProg', 'Car3rd', 'CPA',
            'CarMis', 'CarDis', 'Rec', 'RecProg', 'CrdY', 'CrdR', '2CrdY', 'Fls', 'Fld', 'Off', 'Crs', 'TklW', 'PKwon', 'PKcon', 'OG', 
            'Recov', 'AerWon', 'AerLost', 'AerWon%']

# Coppy the Data Frame from PL_complet
PL_complet_def_atack = PL_complet[columnas].copy()

# Here I define the Ofensive staistics of the team, They will be sum  some other will be avg.
estadisticas_ofensivas_PL = PL_complet_def_atack.groupby('Squad').agg({
    'Goals': lambda x: x.sum() * 0.15,
    'Shots': lambda x: x.sum() * 0.1,
    'G/Sh': lambda x: x.mean() * 0.03,
    'Assists': lambda x: x.sum() * 0.1,
    'SCA': lambda x: x.sum() * 0.08,
    'ScaPassLive': lambda x: x.sum() * 0.05,
    'ScaPassDead': lambda x: x.sum() * 0.04,
    'ScaDrib': lambda x: x.sum() * 0.06,
    'ScaSh': lambda x: x.sum() * 0.06,
    'ScaFld': lambda x: x.sum() * 0.03,
    'GCA': lambda x: x.sum() * 0.07,
    'GcaPassLive': lambda x: x.sum() * 0.07,
    'GcaPassDead': lambda x: x.sum() * 0.05,
    'GcaDrib': lambda x: x.sum() * 0.07,
    'GcaSh': lambda x: x.sum() * 0.06,
    'GcaFld': lambda x: x.sum() * 0.05,
    'PKatt': lambda x: x.sum() * -0.04,
    'PKwon': lambda x: x.sum() * 0.08,
    'PKcon': lambda x: x.sum() * -0.09,
    'OG': lambda x: x.sum() * -0.15
})


# I do the same with the defensive stats
estadisticas_defensivas_PL = PL_complet_def_atack.groupby('Squad').agg({
    'Tkl': lambda x: x.sum() * 0.08,
    'TklWon': lambda x: x.sum() * 0.1,
    'TklDef3rd': lambda x: x.sum() * 0.04,
    'TklMid3rd': lambda x: x.sum() * 0.05,
    'TklAtt3rd': lambda x: x.sum() * 0.08,
    'TklDri': lambda x: x.sum() * 0.06,
    'TklDriAtt': lambda x: x.sum() * -0.06,
    'TklDri%': lambda x: x.mean()*0.05,
    'TklDriPast': lambda x: x.sum() * 0.04,
    'Blocks': lambda x: x.sum() * 0.07,
    'BlkSh': lambda x: x.sum() * 0.1,
    'BlkPass': lambda x: x.sum() * 0.06,
    'Int': lambda x: x.sum() * 0.09,
    'Tkl+Int': lambda x: x.sum() * 0.1,
    'Clr': lambda x: x.sum() * 0.07,
    'Err': lambda x: x.sum() * -0.08,
    'Recov': lambda x: x.sum() * 0.06,
    'AerWon': lambda x: x.sum() * 0.4,
    'AerLost': lambda x: x.sum() * -0.4,
    'AerWon%': lambda x: x.mean()*0.1,
})

estadisticas_ofensivas_PL_f=estadisticas_ofensivas_PL*0.6
estadisticas_defensivas_PL_f=estadisticas_defensivas_PL*0.4
# Combinar las estadísticas ofensivas y defensivas ajustadas en un nuevo DataFrame
probabilidades_ganar = pd.concat([estadisticas_ofensivas_PL_f, estadisticas_defensivas_PL_f], axis=1)

# Calcular la columna de probabilidad de ganar sumando las estadísticas ajustadas
probabilidades_ganar['Probabilidad Ganar'] = probabilidades_ganar.sum(axis=1) + probabilidades_ganar.sum(axis=1) 

# Ordenar el DataFrame en orden descendente por la columna de probabilidad de ganar
probabilidades_ganar = probabilidades_ganar.sort_values('Probabilidad Ganar', ascending=False)

# Imprimir la tabla en orden descendente de probabilidad de ganar
print("Tabla de probabilidades de ganar:")
print(probabilidades_ganar)
```
